```python
import csv
import numpy as np
from pysmiles import read_smiles
import networkx as nx
from matplotlib import pyplot as plt
import warnings
from tqdm import tqdm
warnings.filterwarnings("ignore")
```

```python
smiles = []
labels = []
with open('data_good/train.csv', newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='|')
    for row in spamreader:
        if '[se]' not in row[1]:
            smiles.append(row[1])
            labels.append(row[2])
smiles = smiles[1:-1]
labels = labels[1:-1]
        
def convert_to_bool(label):
    if label == 'False':
        return False
    if label == 'True':
        return True
    raise ValueError("bad")
    
labels = [convert_to_bool(label) for label in labels]

labels = np.array(labels)

print(np.sum(labels))

mols = [read_smiles(smile) for smile in smiles]
```

```python
spectrums = [nx.linalg.spectrum.laplacian_spectrum(mol)[::-1] for mol in mols]
```

```python
print(len(spectrums))
print(len(labels))
```

```python
def convert(spectrum, num):
    if len(spectrum) >= num:
        return spectrum[:num]
    else:
        result = np.zeros(num)
        result[:len(spectrum)] = spectrum
        return result
    
converted = [convert(spectrum, 50) for spectrum in spectrums]
converted = np.array(converted)
print(converted.shape)
```

```python
all_symbols = []
for mol in mols:
    for data in mol.nodes(data = 'element'):
        all_symbols.append(data[1])
all_symbols = np.unique(all_symbols)
print(all_symbols)
```

```python
compositions = []
for mol in mols:
    compositions_now = np.zeros(len(all_symbols))
    for data in mol.nodes(data = 'element'):
        for i in range(len(all_symbols)):
            if all_symbols[i] == data[1]:
                compositions_now[i] += 1
    compositions.append(compositions_now)
compositions = np.array(compositions)
print(compositions.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import f1_score
```

```python
for multiplier in [0, 0.000001, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000, 100000]:
    clf = RandomForestClassifier(n_estimators = 1000, n_jobs = -1)
    weights = np.ones(labels.shape)
    weights[labels] *= multiplier
    predictions = cross_val_predict(clf, np.concatenate([compositions, converted], axis = 1), 
                                    labels, method = 'predict_proba', fit_params={'sample_weight': weights})[:, 1]

    values = []
    for threshold in np.linspace(0, 1, 1000):
        values.append(f1_score(labels, predictions > threshold))
    print(multiplier, np.max(values))
```
